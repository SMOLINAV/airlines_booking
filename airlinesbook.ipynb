```python
# Importar las librerías necesarias
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Guardar en la variable 'ruta' la url del dataset
ruta = "https://raw.githubusercontent.com/SMOLINAV/dataair/main/customer_booking%20-%20customer_booking.csv"

data = pd.read_csv(ruta, sep=",")

print(data.dtypes)  # Mostrar los tipos de datos en el DataFrame data
```

```python
# Crear un histograma de la duración de los vuelos
plt.figure(figsize=(10, 6))
plt.hist(data['flight_duration'], bins=20, color='red', edgecolor='black')
plt.title('Distribución de Duración de Vuelos')
plt.xlabel('Duración del Vuelo (horas)')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
# Convertir el día de la semana a valores numéricos
day_of_week_mapping = {'Mon': 1, 'Tue': 2, 'Wed': 3, 'Thu': 4, 'Fri': 5,
                       'Sat': 6, 'Sun': 7}
data['flight_day_numeric'] = data['flight_day'].map(day_of_week_mapping)

# Crear un gráfico de dispersión de duración de vuelos vs. día de la semana
plt.figure(figsize=(10, 6))
plt.scatter(data['flight_day_numeric'], data['flight_duration'], alpha=0.5,
            color='red')
plt.xticks(list(day_of_week_mapping.values()), list(day_of_week_mapping.keys()))
plt.title('Duración de Vuelo vs. Día de la Semana')
plt.xlabel('Día de la Semana')
plt.ylabel('Duración del Vuelo (horas)')
plt.grid(True)
plt.show()
```

```python
# Grafica distribución de numeros de pasajeros
plt.figure(figsize=(10, 6))
sns.histplot(data['num_passengers'], bins=30, kde=True, color='red')
plt.title('Distribution of Number of Passengers')
plt.xlabel('Number of Passengers')
plt.ylabel('Frequency')
plt.show()
```

```python
# Grafica de distribución de canales de venta
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='sales_channel', data=data)
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                 (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha = 'center', color='green', va = 'center',
                 xytext = (0, 5),
                 textcoords = 'offset points')

plt.title('Distribution of Sales Channels')
plt.xlabel('Sales Channel')
plt.show()
```

```python
# Grafica la distribución de las duraciones de vuelo
plt.figure(figsize=(10, 6))
sns.histplot(data['flight_duration'], bins=30, kde=True, color='green')
plt.title('Distribución de las Duraciones de Vuelo')
plt.xlabel('Duración del Vuelo (horas)')
plt.ylabel('Frecuencia')
plt.show()
```
